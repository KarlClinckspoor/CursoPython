```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib
```

```python
x = np.linspace(0, 10, num=1000)

#y_manual = 0.3 * x ** 2
y_auto = 3 * (x - 1)
y_repeat = 3 * x
y_auto = np.clip(y_auto, 0, np.inf)

fig, ax = plt.subplots()

# Creating Bezier curves for manual and repeat
Path = mpath.Path
pts_manual = matplotlib.bezier.find_control_points(0, 0, 1, 0.5, 3, 2)
manual1 = mpatches.PathPatch(
    Path(pts_manual, [Path.MOVETO, Path.CURVE3, Path.CURVE3]),
    fc='None', ec='C0', transform=ax.transData
)
pts_manual_repeat = matplotlib.bezier.find_control_points(0, 0, 1, 1, 3, 2.5)
manual2 = mpatches.PathPatch(
    Path(pts_manual_repeat, [Path.MOVETO, Path.CURVE3, Path.CURVE3]),
    fc='None', ec='C2', transform=ax.transData
)

p1 = ax.add_patch(manual1)
p2 = ax.add_patch(manual2)

#plt.plot(x, y_manual, label='Manual')
p3 = ax.plot(x, y_auto, c='C1', label='Automação')
p4 = ax.plot(x, y_repeat, c='C3', label='Repetição da automação')
ax.set_ylim(-0.05, 10)
ax.set_xlim(0, 3)
ax.set_xticks([])
ax.set_yticks([])

ax.legend([p1, p3[0], p2, p4[0]], ["Manual", "Automação", "Repetição do manual", "Repetição da automação"])
```
