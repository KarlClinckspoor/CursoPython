```python
# Benzeno com círculo
import turtle
turtle.resizemode('auto')
length = 50
sides = 6
angle = 360 / sides
distance_to_double_bond = length / 10

for edge in range(sides):
    turtle.forward(length)
    if edge % 2 == 0:
        # Armazenar posição atual
        xpos, ypos = turtle.position()
        heading = turtle.heading()

        # Posicionar no local da dupla ligação
        turtle.penup()
        turtle.right(90)
        turtle.forward(distance_to_double_bond)
        turtle.right(90)

        # Desenhar dupla ligação
        turtle.pendown()
        turtle.forward(length)

        # Voltar à posição inicial
        turtle.penup()
        turtle.setposition(xpos, ypos)
        turtle.setheading(heading)
        turtle.pendown()
    
    turtle.right(angle)

turtle.done()
```

```python
# "Benzino"
import turtle
turtle.resizemode('auto')
length = 50
sides = 6
angle = 360 / sides
distance_to_double_bond = length / 10

for edge in range(sides):
    turtle.forward(length)
    if edge % 2 == 0:
        # Armazenar posição atual
        xpos, ypos = turtle.position()
        heading = turtle.heading()

        # Posicionar no local da dupla ligação
        turtle.penup()
        turtle.right(90)
        turtle.forward(distance_to_double_bond)
        turtle.right(90)

        # Desenhar dupla ligação
        turtle.pendown()
        turtle.forward(length)

        # Ir para outro lado
        turtle.penup()
        turtle.right(90)
        turtle.forward(2*distance_to_double_bond)
        turtle.right(90)
        
        # Desenhar outra ligação
        turtle.pendown()
        turtle.forward(length)

        # Voltar à posição inicial
        turtle.penup()
        turtle.setposition(xpos, ypos)
        turtle.setheading(heading)
        turtle.pendown()
    
    turtle.right(angle)

turtle.done()
```

```python
# Benzeno com círculo, encontrando lado oposto mais facilmente
import turtle

turtle.resizemode('auto')
length = 50
sides = 6
angle = 360 / sides
distance_to_double_bond = length / 10

starting_pos = turtle.position()
opposite_side_pos = None
prev_heading = turtle.heading()

for edge in range(sides):
    turtle.forward(length)
    turtle.right(angle)

    curr_heading = turtle.heading()
    if (turtle.heading() == 180) or ((curr_heading < 180) and (prev_heading > 180)) :
        opposite_side_pos = turtle.position()
    prev_heading = curr_heading

# Calcular centro do círculo
delta_h = abs(starting_pos[1] + opposite_side_pos[1])
circle_center = (length / 2, - delta_h / 2)
circle_radius = 0.35 * delta_h
offset = (delta_h - circle_radius) / 2

# Desenhar círculo
turtle.penup()
turtle.setposition(circle_center)
turtle.setheading(90)
turtle.forward(circle_radius)
turtle.right(90)
turtle.pendown()
turtle.circle(-circle_radius)
turtle.penup()

turtle.done()
```
