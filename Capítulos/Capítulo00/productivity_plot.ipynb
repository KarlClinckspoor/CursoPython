```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D


x = np.linspace(0, 10, num=1000)

y_auto = np.clip(3 * (x - 1), 0, np.inf)
y_repeat = np.clip(3 * (x - 0.1), 0, np.inf)


fig, ax = plt.subplots()

# Creating Bezier curves for manual and repeat
Path = mpath.Path
pts_manual = matplotlib.bezier.find_control_points(0, 0, 1, 0.35, 3, 2)
manual1 = mpatches.PathPatch(
    Path(pts_manual, [Path.MOVETO, Path.CURVE3, Path.CURVE3]),
    fc='None', ec='C0', transform=ax.transData
)
pts_manual_repeat = matplotlib.bezier.find_control_points(0, 0, 1, 1, 3, 2.5)
manual2 = mpatches.PathPatch(
    Path(pts_manual_repeat, [Path.MOVETO, Path.CURVE3, Path.CURVE3]),
    fc='None', ec='C2', transform=ax.transData
)

p1 = ax.add_patch(manual1)
p2 = ax.add_patch(manual2)

#plt.plot(x, y_manual, label='Manual')
p3 = ax.plot(x, y_auto, c='C1', label='AutomaÃ§Ã£o', ls='--')
p4 = ax.plot(x, y_repeat, c='C3', label='RepetiÃ§Ã£o da automaÃ§Ã£o', ls='--')
ax.set_ylim(-0.05, 10)
ax.set_xlim(0, 3)
ax.set_xticks([])
ax.set_yticks([])

p_manual = Line2D([0, 1], [0, 1], c='C0')
p_manual_repeat = Line2D([0, 1], [0, 1], c='C2')

ax.legend([p_manual, p3[0], p_manual_repeat, p4[0]], ["Manual", "AutomaÃ§Ã£o", "RepetiÃ§Ã£o do manual", "RepetiÃ§Ã£o da automaÃ§Ã£o"])

ax.annotate(
    'LatÃªncia', 
    xy=(0.2, 0.09),
    xytext=(0.135, 0.02),
    xycoords='figure fraction', 
    arrowprops=dict(arrowstyle='-[, widthB=6, angleB=22, lengthB=0.8')
)

```
